```python
import pandas as pd
from yahooquery import Ticker
```

```python
df_list = pd.read_csv('data/stock_list.csv')
df_list
```

```python
df_industry = df_list[(df_list['Sector'] == 'Utilities') & (df_list['Country'] == 'USA')]
df_industry
```

```python
tickers = Ticker(df_industry['Ticker'].values, asynchronous=True)

# Default period = ytd, interval = 1d
df_stock = tickers.history(start='2023-03-01',end='2024-05-01',interval='1d')

df_stock
```

```python
df_prices = df_stock['adjclose'].unstack(level=0)
df_prices.index = pd.to_datetime(df_prices.index)
df_prices.dropna(axis=1,inplace=True)
df_prices
```

```python
df_prices.to_csv('stock_prices.csv')
```
