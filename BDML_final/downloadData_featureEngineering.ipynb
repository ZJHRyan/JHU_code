```python
import pandas as pd
from yahooquery import Ticker
from talib.abstract import *

import warnings
warnings.filterwarnings('ignore')
```

```python
stock_list = pd.read_csv('stock_list.csv')
```

```python
df_etf = stock_list[stock_list['Market Cap'] != '-']
df_etf
```

```python
# Function to convert values
def convert_to_millions(value):
    if 'B' in value:
        return float(value.replace('B', '')) * 1000
    elif 'M' in value:
        return float(value.replace('M', ''))
    else:
        return float(value)

# Apply the function to the 'Market Cap' column
df_etf['Market Cap'] = df_etf['Market Cap'].apply(convert_to_millions)
df_etf
```

```python
# Calculate the 10th percentile of MarketCap
percentile_10 = df_etf['Market Cap'].quantile(0.05)

# Filter the DataFrame to get rows where MarketCap is less than the 10th percentile
filtered_df = df_etf[df_etf['Market Cap'] < percentile_10]

filtered_df
```

```python
tickers = Ticker(filtered_df['Ticker'].values, asynchronous=True)

# Default period = ytd, interval = 1d
df_stock = tickers.history(start='2022-05-01',end='2024-05-01',interval='1d')

df_stock
```

```python
# Convert 'date' in MultiIndex to Timestamp
df_stock.index = df_stock.index.set_levels(pd.to_datetime(df_stock.index.levels[1]), level='date')
df_stock
```

```python
# df_stock.to_csv('df_stock.cvs')
```

```python
columns_index = [
    # Overlap Studies
    'DEMA', 'EMA', 'HT_TRENDLINE',
    'KAMA', 'MA', 'MIDPOINT',
    'MIDPRICE', 'SAR', 'SAREXT',
    'SMA', 'T3', 'TEMA', 'TRIMA', 'WMA',
    # Momentum Indicators
    'ADX', 'ADXR', 'APO',
    'AROONOSC', 'BOP',
    'CCI', 'CMO', 'DX',
    'MFI', 'MINUS_DI', 'MINUS_DM',
    'MOM', 'PLUS_DI', 'PLUS_DM',
    'PPO', 'ROC', 'ROCP',
    'ROCR', 'ROCR100', 'RSI',
    'TRIX', 'ULTOSC', 'WILLR',
    # Volume Indicators
    'AD', 'OBV', 'ADOSC',
    # Cycle Indicators
    'HT_DCPERIOD', 'HT_DCPHASE',
    # Price Transform
    'AVGPRICE', 'MEDPRICE', 'TYPPRICE', 'WCLPRICE',
    # Volatility Indicators
    'ATR', 'NATR', 'TRANGE',]
    # # Statistic Functions
    # 'BETA', 'CORREL', 'LINEARREG', 'LINEARREG_ANGLE',
    # 'LINEARREG_INTERCEPT', 'LINEARREG_SLOPE',
    # 'STDDEV', 'TSF', 'VAR']


def convert_df(x, df):
    inputs = {
        'open': df["open"].to_numpy(),
        'high': df["high"].to_numpy(),
        'low': df["low"].to_numpy(),
        'close': df["adjclose"].to_numpy(),
        'volume': df["volume"].to_numpy()
    }
    y = eval(x)
    df[x] = 0.0
    df[x] = y(inputs)
    return df


def calculate_index(df):
    for i in columns_index:
        convert_df(i, df)
    return df
```

```python
df_techIndi = df_stock.groupby(['symbol'])[
    ['open', 'high', 'low', 'adjclose', 'volume']].apply(lambda x: calculate_index(x))
df_techIndi
```

```python
df_techIndi.dropna(inplace=True)
df_techIndi
```

```python
df_techIndi.to_csv('df_techIndi.csv')
```

# Merge data

```python
df_techIndi['RET'] = df_techIndi.groupby(['symbol'])['adjclose'].apply(lambda x: x/x.shift(1)-1).values
df_select = df_techIndi[['RET','adjclose','MIDPOINT','ADX','MINUS_DM','PLUS_DI','PPO']]
df_select
```

```python
df_select.reset_index(inplace=True)
df_select.set_index(['date'],inplace=True)
df_select
```

```python
df_five_factor = pd.read_csv('F-F_Research_Data_5_Factors_2x3_daily.csv',index_col='date', parse_dates=True)
df_five_factor = df_five_factor.drop('RF',axis=1)
df_five_factor
```

```python
merged_df = pd.merge(df_select, df_five_factor, on='date', how='inner')
merged_df
```

```python
merged_df.dropna(inplace=True)
merged_df.to_csv('merged_df.csv')
```
